```python
# prompt: grafica la media movil del año 2015 y el precio de close y que los valores en x se muestren cada por cada mes

import pandas as pd
import matplotlib.pyplot as plt



ms_2015 =  ms[(ms['Date'] >= '2015-01-01')&( ms['Date']<= '2015-12-31')]
ms_2015['Date'] = pd.to_datetime(ms_2015['Date']) # Convert 'Date' column to datetime objects
ms_2015 = ms_2015.set_index('Date') # Set 'Date' as index for easier plotting

plt.figure(figsize=(14,7))
plt.plot(ms_2015['Close'], label='Close Price')
plt.plot(ms_2015['ma60'], label='60-Day Moving Average')

plt.xlabel('Month of 2015')
plt.ylabel('Price')
plt.title('Microsoft Stock Price and 60-Day Moving Average (2015)')
plt.legend()



plt.show()
```

```python
ms['MA10']=ms['Close'].rolling(10).mean()
ms['MA50']=ms['Close'].rolling(50).mean()
ms = ms.dropna()
ms.head()
```

```python
ms['shares'] = ms.apply(lambda x: 1 if x['MA10']  > x['MA50'] else 0, axis=1)
```

```python
ms['Close1'] = ms['Close'].shift(-1)
ms['profit'] = ms.apply(lambda x : x['Close1'] - x['Close'] if x['shares'] == 1 else 0, axis=1)
ms['profit'].plot()
plt.axhline(0, color='r')
```

```python
ms['wealth']=ms['profit'].cumsum()
ms = ms.dropna()
ms.tail()
```

```python
ms['wealth'].plot()
plt.title('Total money you win is {}'.format(ms['wealth'].iloc[-1]))
```
